```python
factors_data = data['Which of these factors are important to you when deciding which car to purchase?']

# Extract the column containing factors
factors_data = data['Which of these factors are important to you when deciding which car to purchase?']

# Handle NaN values by converting them to an empty string
factors_data.fillna('', inplace=True)

# Split factors for each respondent and count occurrences
factor_counts = {}
for factors_str in factors_data:
    factors = factors_str.split(';')
    for factor in factors:
        if factor.strip() in factor_counts:
            factor_counts[factor.strip()] += 1
        else:
            factor_counts[factor.strip()] = 1

# Plot
plt.figure(figsize=(12, 6))
plt.title('Distribution of Factors Influencing Car Purchase Decision')
sns.barplot(x=list(factor_counts.keys()), y=list(factor_counts.values()), palette="muted")
plt.xlabel('Factors')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()

# Display counts on the plot
for i, v in enumerate(factor_counts.values()):
    plt.text(i, v + 0.5, f'{v}%', ha='center')

plt.show()
```

```python
from scipy.stats import chi2_contingency

# The observed frequencies from your table
observed_frequencies = np.array([[1, 8, 32],
                                 [0, 2, 2],
                                 [0, 1, 2],
                                 [0, 1, 1]])

# Perform the chi-square test to get the chi-square value
chi2, _, _, _ = chi2_contingency(observed_frequencies)

# Calculate Cramér's V
num_rows, num_cols = observed_frequencies.shape
cramers_v = np.sqrt(chi2 / (np.sum(observed_frequencies) * (min(num_rows, num_cols) - 1)))

# Print the results
print(f"Chi-square value: {chi2}")
print(f"Cramér's V value: {cramers_v}")
```
