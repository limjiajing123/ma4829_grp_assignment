# Age group and car customisation correlation

```python
# Drop rows with missing values in the relevant columns
data_cleaned = data.dropna(subset=['Which age group do you belong to?',
                                'How likely are you to opt for customised vehicle if there were no extra charges? '])

# data_cleaned

# Convert the categorical variable to numerical values (for ordinal data)
# You can assign numerical values based on the order of likelihood (e.g., Very likely = 3, Likely = 2, Not likely = 1)
# Adjust the mapping based on your specific data
mapping = {'Not likely': 1, 'Likely': 2, 'Very likely': 3}
data_cleaned['Likelihood_numerical'] = data_cleaned['How likely are you to opt for customised vehicle if there were no extra charges? '].map(mapping)

# Use the 'pd.crosstab()' function to create a contingency table
contingency_table = pd.crosstab(data_cleaned['Which age group do you belong to?'], data_cleaned['Likelihood_numerical'])

print(" order of likelihood (e.g., Very likely = 3, Likely = 2, Not likely = 1)")
contingency_table
```

```python
from scipy.stats import chi2_contingency

# The observed frequencies from your table
observed_frequencies = np.array([[1, 8, 32],
                                 [0, 2, 2],
                                 [0, 1, 2],
                                 [0, 1, 1]])

# Perform the chi-square test to get the chi-square value
chi2, _, _, _ = chi2_contingency(observed_frequencies)

# Calculate Cramr's V
num_rows, num_cols = observed_frequencies.shape
cramers_v = np.sqrt(chi2 / (np.sum(observed_frequencies) * (min(num_rows, num_cols) - 1)))

# Print the results
print(f"Chi-square value: {chi2}")
print(f"Cramr's V value: {cramers_v}")
```

```python
# The observed frequencies from your table
observed_frequencies = np.array([[1, 8, 32],
                                 [0, 2, 2],
                                 [0, 1, 2],
                                 [0, 1, 1]])

# Convert observed frequencies to DataFrame for better visualization
columns = ['1', '2', '3']  # Likelihood levels
index = ['20-30', '31-40', '41-50', '51-60']  # Age groups
df_observed = pd.DataFrame(observed_frequencies, columns=columns, index=index)

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df_observed, annot=True, cmap="Blues", fmt="d", cbar=False)
plt.title("Observed Frequencies Heatmap")
plt.xlabel("Likelihood Numerical")
plt.ylabel("Age Group")
plt.show()
```

```python
# Categories and Likelihood levels
categories = ['20-30', '31-40', '41-50', '51-60']
likelihood_levels = ['Not likely', 'Likely', 'Very likely']

# Plotting the bar chart
width = 0.2  # Width of the bars
bar_positions = np.arange(len(categories))

plt.figure(figsize=(10, 6))

for i, level in enumerate(likelihood_levels):
    plt.bar(bar_positions + i * width, observed_frequencies[:, i], width=width, label=f'Level {level}')

plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.title('Distribution of Likelihood Levels by Age Group')
plt.xticks(bar_positions + width, categories)
plt.legend(title='Likelihood Levels')
plt.show()
```

```python
# Splitting multiple choices into separate rows
data['Which of the following exterior components would you choose to customise (texture, layout, size, etc)? '] = data['Which of the following exterior components would you choose to customise (texture, layout, size, etc)? '].str.split(';')
data = data.explode('Which of the following exterior components would you choose to customise (texture, layout, size, etc)? ')

# Creating a contingency table
contingency_table = pd.crosstab(data['Which of the following best describes you?'], data['Which of the following exterior components would you choose to customise (texture, layout, size, etc)? '])

# Performing chi-square test of independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Number of observations
n = data.shape[0]

# Calculate Cramer's V
cramer_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))

print("Cramer's V:", cramer_v)
print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(expected)

contingency_table
```

# From YES for personalised design, how much are they willing to spend on personalised design?

```python
data_cleaned = data.dropna(subset=['Are you interested in designing your own components to personalise your car? ',
                                'How much are you willing to pay for the personalised design? '])

data_cleaned['Design Interest'] = data_cleaned ['Are you interested in designing your own components to personalise your car? ']
data_cleaned['Price Range'] = data_cleaned ['How much are you willing to pay for the personalised design? ']

# Step 1: Create contingency table
contingency_table = pd.crosstab(data_cleaned['Design Interest'], data_cleaned['Price Range'])

# Step 2: Calculate chi-square statistic and Cramer's V
chi2, p, dof, expected = chi2_contingency(contingency_table)
n = contingency_table.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))

# Step 3: Filter data for responses where individuals are interested in designing their own components
interested_responses = data_cleaned[data_cleaned['Design Interest'] == 'Yes']

# Step 4: Calculate percentage distribution of price range for those who said "Yes"
price_range_distribution = interested_responses['Price Range'].value_counts(normalize=True) * 100

# Step 5: Present results in a table
print("Cramer's V:", cramers_v)
print("\nPercentage Distribution of Price Range for Individuals Interested in Designing Their Own Components:")
print(price_range_distribution)

plt.figure(figsize=(8, 6))
price_range_distribution.plot(kind='bar', color='skyblue')
plt.title('Price Range Distribution for Individuals Interested in Designing Their Own Components')
plt.xlabel('Price Range')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
