```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.preprocessing import OneHotEncoder
```

```python
pd.set_option('display.max_rows', None)
```

```python
file_path = 'SURVEY RESULTS_2024.xlsx'
df = pd.read_excel(file_path)
df.head()
```

```python
columns = df.columns
for i, column in enumerate(columns):
    print(column)
```

```python
renamed_labels={1: "Age group",
                2: "Gender",
                3: "Car ownage",
                4: "Marital status",
                5: "Car factors",
                6: "Car customization",
                7: "Exterior components",
                8: "Interior components",
                9: "Spending on customisation",
                10: "Design personalised car",
                11: "Spending on personalised design",
                12: "3D experience",
                13: "Design ideas"
               }
```

```python
for i, column in enumerate(columns):
    df.rename(columns = {column:renamed_labels[i+1]}, inplace=True)
new_columns = df.columns
df
```

```python
df['Car ownage'] = df['Car ownage'].replace('Do not own a car, but planning to purchase in future', 'Plan to buy')
df['Gender'] = df['Gender'].replace('Prefer not to say', 'Gender unknown')
df['Design personalised car'] = df['Design personalised car'].replace('Only with professional help', 'Professional help')
df['Spending on customisation'] = df['Spending on customisation'].astype(str)
df['Spending on personalised design'] = df['Spending on personalised design'].astype(str)
```

```python
age_A = df[[column for i, column in enumerate(new_columns) if i in [0,1,2,3,5,8]]]
age_A
```

```python
encoder = OneHotEncoder(sparse=False)
encoded_data_A = encoder.fit_transform(age_A)
column_names = [val for sublist in encoder.categories_ for val in sublist]
encoded_df_A = pd.DataFrame(encoded_data_A, columns=column_names)
encoded_df_A
```

```python
frequent_itemsets_A = apriori(encoded_df_A, min_support=0.2, use_colnames=True)
frequent_itemsets_A
```

```python
likelihood = ['Likely', 'Very likely']
itemsets_A = frequent_itemsets_A[frequent_itemsets_A['itemsets'].apply(lambda x: any(item in x for item in likelihood))]
itemsets_A['length'] = itemsets_A['itemsets'].apply(lambda x: len(x))
itemsets_A[(itemsets_A['length'] >=2)]
```

```python
rules_A = association_rules(frequent_itemsets_A, metric="confidence", min_threshold=0.1)
rules_A = rules_A[rules_A['consequents'].apply(lambda x: any(item in x for item in likelihood) and len(x)<2)]
rules_A = rules_A[rules_A['antecedents'].apply(lambda x: len(x) == 1)]
rules_A = rules_A.drop(columns=['leverage', 'zhangs_metric', 'conviction'])
rules_A['consequents_str'] = rules_A['consequents'].apply(lambda x: ', '.join(sorted(x)))
rules_A = rules_A.sort_values(by='consequents_str', ascending=True)
rules_A = rules_A.drop(columns=['consequents_str'])
rules_A
```
